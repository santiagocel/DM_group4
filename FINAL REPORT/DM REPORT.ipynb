## 1.Introduction

E-Commerce websites have become a popular strategy for many businesses to seamlessly mar-ket and sell their products to customers. It is crucial that these websites operate smoothly tocaptivate more customers and attain competitive advantage and the platforms are heavilyreliant on proper data management. This report discusses in detail the data managementaspects of “We-Buy” a simulated E-Commerce environment from database design to analysisand reporting.

## 2.Database Design and Implementation

**2.1 Requirement Collection and Analysis**  

**2.1.1. Functional Requirements**  

-   Customers can buy products offered by Sellers online 

-   Customer can place an order for products bought and make payments 

-   Customer order will be delivered, and transaction will be complete 

-   Supplier Records will be updated with the product details 

-   Products will be categorized, and Ads can be created for products 

-   All relevant entities must be updated when a record is entered  

-   User details must be treated with confidentiality and Database must be secure

**2.1.2. Data Requirements** 

**2.1.2.1 Data Entities and Attributes** 

Customer Details 

-   Entity: Customer 

-   Attributes: Customer ID, First name, Last name, Email, Password, Gender, Membership, DOB, Customer address  

Order Details  

-   Entity: Order  

-   Attributes: Order ID, Customer ID, Product ID, Order date, Order status, Sub quantity, Total quantity, Total price 

Transaction Details  

-   Entity: Transaction 

-   Attributes: Transaction ID, Order ID, Customer ID, Transaction time, Transaction status, Amount, Payment method 

Delivery Details  

-   Entity: Delivery  

-   Attributes: Delivery ID, Transaction ID, Delivery start date, Delivery end date, Delivery status 

Product Details  

-   Entity: Product 

-   Attributes: Product ID, Category ID, Product price, Purchase price, Product name, Discount price, Rate value 

Supply Details  

-   Entity: Supply  

-   Attributes: Supplier ID, Product ID, Quantity available 

Supplier Details  

-   Entity: Supplier  

-   Attributes: Supplier ID, Supplier email, Supplier name, Supplier phone, Supplier Address 

Ads Details  

-   Entity: Ads 

-   Attributes: Ad ID, Product ID, Ad name 

Category Details  

-   Entity: Category  

-   Attributes: Category ID, Parent category ID, Category type 

**2.1.2.2 Data Relationships**

One to One relationship 

-   Entity: Customer and Transaction 

-   Relationship: A Customer can have only one transaction at a time, and one transaction will be entered in to by one customer.
